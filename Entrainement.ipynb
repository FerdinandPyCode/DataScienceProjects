```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn
```

```python
tab=np.arange(0,1000,5)
```

```python
tab.shape
```

```python
tab.reshape(10,20)
tab
```

```python
y=tab**2-tab-100
```

```python
plt.figure(figsize=(12,12))
plt.xlabel('Un petit essai en X')
plt.ylabel('Un autre essai en Y')
plt.scatter(tab,y,c='r',label='forme')
plt.scatter(tab,tab**2-200,c='green',label='Another Forme')
plt.legend()
plt.show()
```

```python
np.random.seed(0)
m=100
x=np.linspace(0,10,m).reshape(m,1)
y=x+np.random.randn(m,1)
plt.scatter(x,y)
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x,y)
model.score(x,y)
prediction=model.predict(x)
plt.scatter(x,y,label='Depart')
plt.plot(x,prediction,c='r',label='Regression')
plt.legend()
plt.xlabel('X')
plt.ylabel('Y')
```
