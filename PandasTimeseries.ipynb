```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('BTC-EUR.csv',index_col='Date',parse_dates=True)
```

```python
bitcoin=data
```

```python
bitcoin.head()
```

```python
bitcoin['Close'].plot(figsize=(9,6))
plt.show()
```

```python
bitcoin.loc['2019','Close'].resample('M').plot()
```

```python
bitcoin.loc['2019','Close'].resample('2M').mean().plot()
```
